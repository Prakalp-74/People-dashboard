```python
highest_salary = df.groupby("salary")["dept"].max()
```

```python
satisfaction_low = df.groupby("satisfactoryLevel")["dept"].min()
```

```python
satisfaction_high = df.groupby("satisfactoryLevel")["dept"].max()
```

```python
low_salary = df.groupby("dept")["salary"]
```

```python
dept_project = df.groupby("dept")["numberOfProjects"]
```

```python
dept_last_ev = df.groupby("lastEvaluation")["dept"]
```

```python
dept_wise_left = df.groupby("dept")["left"]
```

```python
print("highest salary", highest_salary)
print("satisfacation level low", satisfaction_low)
print("satisfaction level high", satisfaction_high)
print("lower salary", low_salary)
print("dept wise projects", dept_project)
print("department wise last evalution",dept_last_ev)
print("department wise left", dept_wise_left)
```
